```python
def collaborative_rec(user):
    if type(user) == int:
        user = ratings.loc[user]
    movies_not_seen = user.isna()
    nearest_neighbors = NearestNeighbors(n_neighbors=20).fit(ratings.head(100).fillna(ratings.mean()))
    nearest_neighbors = nearest_neighbors.kneighbors(np.array([user.fillna(user.mean())]), return_distance=False)
    best_movies = ratings.iloc[nearest_neighbors[0]]
    best_movies = best_movies[best_movies.columns[user.isna()]]
    best_movies = best_movies.dropna(axis=1, how='all')
    best_movies = best_movies.mean().sort_values(ascending=False)
    return best_movies[:10].index

```
