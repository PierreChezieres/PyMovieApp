```python
import pandas as pd
df = pd.read_csv('top_5000_actors_matrix.csv', header=0, index_col='movieId')
```

```python
df.head()
```

```python
actors = df
movies = pd.read_csv('titles.csv')
```

```python
movies.head()
```

```python
movies = movies.set_index('movieId')
movies.head()
```

```python
actors = pd.merge(left=actors.reset_index(), right=movies.reset_index(), on='movieId')
```

```python
actors.shape()
```

```python
actors
```

```python
actors = actors.set_index('movieId')
```

```python
actors
```

```python
actors = actors.drop('title', axis='columns')
```

```python
actors.to_csv('actors.csv', sep=',')
```

```python
actors = actors.transpose()
```

```python
x, y = actors.shape
actor_list = pd.Series()
for i in range(x):
    actor = actors.iloc[i,:]
    actor = actor.replace({0:pd.NA, 1:actor.name}).dropna()
    actor_list = actor_list.append(actor)
```

```python
actor_list.to_csv('actor_list.csv', sep=',')
```
