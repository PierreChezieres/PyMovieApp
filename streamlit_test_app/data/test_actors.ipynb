```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors
```

```python
def content_based_rec(id, n_rec=10, s='euclidean', m='euclidean'):
    scaler = StandardScaler()
    data = actors.drop(['movieId', 'imdbId', 'tmdbId'], axis=1)
    data_scaled = scaler.fit_transform(data)
    n_neighbors = NearestNeighbors(n_neighbors=n_rec+1, algorithm='auto', metric=m)
    n_neighbors.fit(data_scaled)
    movie = actors[actors['movieId']==id]
    movie_data = scaler.transform(movie.drop(['movieId', 'imdbId', 'tmdbId'], axis=1))
    dist, results = n_neighbors.kneighbors(movie_data, n_rec+1, return_distance=True)
    rec = actors.iloc[results[0][1:],:]
    return rec, dist
```

```python
r, d = content_based_rec(1240)
```
