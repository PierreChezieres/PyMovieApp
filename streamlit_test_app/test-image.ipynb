```python
from PIL import Image
import requests
url = 'https://www.themoviedb.org/t/p/w300_and_h450_bestv2/caX3KtMU42EP3VLRFFBwqIIrch5.jpg'
img = Image.open(requests.get(url, stream=True).raw)
mask = Image.open('mask.png').convert('L')
img = img.resize(mask.size)
img.putalpha(alpha=mask)
```
